```python
import urllib2
opener = urllib2.build.opener()
opener.addheaders = [('User-agent', 'Mozilla/5.0')]

url = "https://www.amerisourcebergen.com/about-who-we-are/"

ourUrl = opener.open(url).read(0)
soup = BeautifulSoup(ourUrl)
title = soup.title.text
body = soup.find(text="innovation").findNext('p')
outfile = open ('innovation_p_tag.txt', 'w')
outfile.write(body.text)
```
